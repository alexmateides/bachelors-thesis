```python
import os
from collections import Counter
import numpy as np

def get_label_counts(data_dir):
    cnt = Counter()

    for filename in os.listdir(data_dir):
        if not filename.endswith(".npz"):
            continue
        path = os.path.join(data_dir, filename)
        with np.load(path, allow_pickle=True) as npz:
            if "label" not in npz.files:
                continue
            label = npz["label"].tolist()
            cnt[label] += 1

    total = sum(cnt.values())
    for label, c in cnt.most_common():
        print(f"{label:<15}\t{c:<8}\t{c/total:.4f}")

print(f"{'Label':<15}\t{'Count':<8}\t{'Fraction'}")
print('-'*36)
get_label_counts("../../data/processed_4_channel_multi_256")
get_label_counts("../../data/processed_4_channel_binary_256")
```
